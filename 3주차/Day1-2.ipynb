## 컨테이너 스토리지 서비스

### Blob Storage

클라우드에 구조화되지 않은 데이터를 저장하는 개체 스토리지입니다.

#### 용도

- 직접 이미지 또는 문서 제공
- 분산 엑세스용 파일 저장
- 비디오 오디오 스트리밍
- 백업, 복원, 재해복구
- 분석 데이터 저장

#### 블롭의 유형

- **Blob**: 일반적인 파일 저장
- **Page Blob**: 읽기/쓰기 위치를 정밀하게 지정해 자주 변경되는 데이터 저장
- **Append Blob**: 계속 누적되는 기록 저장

#### 블롭의 계층

- 핫계층 : 자주 엑세스하거나 수정되는 데이터
- 쿨계층 : 드물게 엑세스되고 최소 30일 동안 저장되는 데이터에 최적화
- 콜드계층 : 90일 이상 드물게 엑세스되거나 또는 수정 및 저장되지 않는 데이터
- 보관 : 검색 대기 시간에 몇 시간 이상씩 걸리며, 최소 180일 이상 보관 계층에 저장되는 데이터
