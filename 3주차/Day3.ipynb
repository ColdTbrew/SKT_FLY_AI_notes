### 가용성이란?

`가용성 = uptime / uptime+downtime`

#### 3가지 다운타임 시나리오

1. 계획되지 않은 하드웨어 유지관리
2. 계획된 유지 관리
3. 예측하지 못한 다운타임

#### 업데이트 도메인

- 배포 전체에서 증분 또는 롤링 업그레이드 수행
- 계획된 유지 관리 동안한한번에 1개의 업데이트 도메인만 다시 부팅

#### 장애 도메인

- 단일 실패 지점을 공유하는 공통 하드웨어 집합인 스위치를 공유하는 그룹
- 가용성 집합의 vm은 두개의 장애도메인에 배치

#### 가용성 집합 - 한 데이타센터에 두개의 컴퓨터에 배포

#### 가용성 영역 - 두개의 다른 영역에 배포 (서울, 부산)

```python
import numpy as np

dict = {
        "최승혁" : 0,
        "김성민" : 1,
        "곽현정" : 2,
        "구나영" : 3,
        "민 건" : 4,
        "김 협" : 5,
        "박민선" : 6,
        "이규호" : 7,
        }

radom_2people_idx = np.random.choice(list(dict.values()), 3, replace=False)
print(radom_2people_idx)
# 이름
radom_2people_name = []
for name, number in dict.items():
    if number in radom_2people_idx:
        radom_2people_name.append(name)

print(radom_2people_name)
```
