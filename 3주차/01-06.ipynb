# 클라우드 교육

#### 클라우드 컴퓨팅이란

- 컴퓨팅
- 네트워킹
- 스토리지
- 분석

##### 특징

- 고가용성
- 내결함성 Fault Tolerence
- 확장성
- 탄력성 Auto scaling
- 글로벌한 범위
- 고객 대기 시간 기능
- 예측 비용 고려 사항 (ML로 예측이 가능해짐)
- 민첩성
- 보안
- 재해 복구

#### 추가 특징

- CapEx 자본 지출, 시간이 지날수록 가치가 낮아짐
- OpEx 운영 비용, 필요에 따라 즉시 청구됨
- 소비 기반 모델, 사용하는 리소스에 대해서만 비용을 지불

### Private Cloud

- 조직의 데이터센터에서 클라우드 환경 구축
- 운영을 조직이 담당
- 외부 엑세스 불가

# IaaS

#### infrastructure as a service

가장 유연한 클라우드 서비스
고객이 하드웨어를 구성하고 관리
