```python
import matplotlib.pyplot as plt 
import seaborn as sns 

# Seaborn 제공 데이터셋 가져오기 
titanic = sns.load_dataset('titanic')

# 스타일 테마 설정 (5가지: darkgrid, whitegrid, dark, white, ticks) 
sns.set_style('ticks') 

# 그래프 객체 생성 (figure에 3개의 서브 플롯을 생성) 
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2) 
ax3 = fig.add_subplot(1, 3, 3)

# distplot 대신 histplot + kde
sns.histplot(titanic['fare'], kde=True, ax=ax1)  # 히스토그램 + KDE
# kdeplot
sns.kdeplot(x='fare', data=titanic, ax=ax2) 
# histplot
sns.histplot(x='fare', data=titanic, ax=ax3)

# 차트 제목 표시 
ax1.set_title('titanic fare - histplot with kde') 
ax2.set_title('titanic fare - kdeplot') 
ax3.set_title('titanic fare - histplot') 

# 레이아웃 조정 및 표시
plt.tight_layout()
plt.show()
```

```python
#씨네21 별점 시각화
import pandas as pd
import requests
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs

res = requests.get('https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&pkid=68&os=33670465&qvt=0&query=모아나%202')
soup = bs(res.content, 'html.parser')

result = soup.find_all(class_="area_text_box_ell")

# print(result)
result2 = soup.find_all(class_='lego_movie_pure_star')

data = []
for name, num in zip(result, result2):
  data.append((name.text, float(num.text)))
# data

df = pd.DataFrame(data, columns=['name', 'rating'])
# df

plt.rc('font', family='NanumBarunGothic')
plt.bar(df['name'], df['rating'])
fig = plt.gcf()
fig.set_size_inches(20,3)
plt.show()
```

```python
#씨네21 별점 시각화
import pandas as pd
import requests
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs
res = requests.get('http://m.cine21.com/movie/review/recent_view/?movie_id=61039')

soup = bs(res.content, 'html.parser')
result = soup.select('.num')
name = soup.select('.name')


names = []

for n in name:
  names.append(n.text)

names

data =[]

for num in result:
  data.append(float(num.text))

data

names

# data : 전문가 평점 / 네티즌 평점 / 등등
df = pd.Series(data)
df

import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.bar(['전문가','네티즌']+[name for name in names],data)
plt.show()
```
