## Docker 컨테이너 수명 주기

---

- 실행 중
  - Dockerfile의 CMD/ENTRYPOINT 인스트럭션에 정의된 앱 실행 중 상태
- 정지
  - 사용자에 의한 명시적 중단
  - 정상적인 실행의 종료
  - 컨테이너 종료 시점의 상태를 디스크에 저장
- 파기
  - 명시적 완전 삭제

## 컨테이너 관리 주요 명령

---

- `docker container run`
  - docker container run 옵션 이미지명
- `docker run -it alpine uname -a`
  - 재정의 명령어
- `docker container run --name guestbook-app -d -p 80:3000 guestbook-app`
  - 포트 열기
  - --filter 필터해서 검색하기
  - -a 종료된 컨테이너 까지 보기
- `docker stop id` 중지
- `docker container restart id` 재시작
- `docker container rm id` 제거
- `docker container **logs -f(표준출력 내용을 계속볼 경우)** <컨테이네D/컨테이너명>`
- `docker exec` -i 표준 입력 연결유지, -t 유사 터미널 할당

```
docker container run -d guestbook-app
docker exec a86f201a7aa7 pwd // pwd 명령어 실행하고 반환만하기
docker exec -it a86f201a7aa7 sh // 도커 안에 들어가서 쉘 열기
docker exec -it a86f201a7aa7 bash // 배시쉘 열기
root@a86f201a7aa7:/app#
```

- `docker cp id 대상파일`
- `docker cp 호스트 파일 대상파일`
- `docker container stats <CONTAINER_ID/CONTAINER_NAME>` 리소스 사용량 확인


## Docker 네트워크

컨네이터가 통신해야함 컨테이너끼리일 수도있고 외부인터넷일 수도 있다.

■ 도커의기본네트워크목록  
`docker network Is`  
■ 네트워크종류  
`bridge(컨테이너 끼리), 
host(로컬 네트워크랑), overlay(여러 호스트사이에 네트워크끼리)
`
■ 네트워크만들기  
`docker network create <net-name>`  
■ 네트워크정보확인  
`docker network inspect <net-name>`  
■ 네트워크에컨테이너 연결  
`docker network connect <net-name> <con-name>`
`docker run -itd -name <con-name> -network <net-name> <img-name>`  
■ 네트워크연결해제  
`docker network disconnect <net-name> <con-name>`  
■ 네트워크제거  
`docker network rm <net-name>`  
■ 핑  
`docker exec firstnode ping secondnode`


## Docker Compose 개요

- 실용적인 시스템 구축
- 애플리케이션간연동
- 하나 이상의 컨테이너가 서로 통신 하며，의존 관계 형성
- 컨테이너 동작 제어를 위한 설정파일이나 환경 변수 전달
- 컨테이너 의존관계를 고려한 포트 포워 딩 설정
- yml으로 docker compose 만들기 위해 준비

## ACR (Azore container registry)

- 지역 복제
- 에저 서비스와 통합이 좋음
  - app service
  - k8s
  - 서비스 패버릭
  - 펑션
  - 배치
- 요금제에 따라 지원되는게 다름 속도, 용량
- CR 배포 (Azure Portal)  
   ① 컨테이너 레지스트리 만들기  
   이름: acr<appname><env><##>  
   SKU: 기본  
   ② 관리사용자활성화  
   ③ 리포지토리확인
