## Dockerfile, Docker Image, Docker Container

---

- 도커파일(명령어 세트)이 빌드되면 이미지(os userland+앱)가되고 생성하면 컨테이너가 됨


#### Docker search

- 도커 허브에 등록된 리포지토리(repo) 검색 `ex. mysql, pytorch`

#### 도커 명령어들

- 도커 허브에서 이미지 가져오기
  `docker pull`

- 도커 이미지들 list하기
  `docker images`

- tag 별명 붙이기
  `docker image tag guestbook:latest steelflea/simple_guestbook:latest`


#### 도커 파일의 구조

- `FROM`: 베이스 이미지를 지정합니다.
- `RUN`: 컨테이너 안에서 명령어를 실행합니다.
- `COPY`: 호스트 시스템의 파일이나 디렉토리를 이미지의 파일 시스템으로 복사합니다.
- `CMD`: 컨테이너가 시작될 때 실행할 명령어를 지정합니다.
- `ADD`: 호스트 시스템의 파일이나 디렉토리를 이미지의 파일 시스템으로 복사합니다. COPY와 유사하지만, URL에서 파일을 다운로드하거나 압축 파일을 자동으로 풀어주는 기능이 추가로 제공됩니다.
- `ENV` : 환경변수 추가
- `EXPOSE` : 열 포트 추가



##### run과 cmd차이

- 런을 빌드 시점에 실행
- cmd은 실행(container)되는 시점에 실행됨
